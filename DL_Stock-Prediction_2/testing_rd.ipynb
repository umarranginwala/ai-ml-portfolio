## LSTM

```python
import yfinance as yf
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
from math import sqrt

# Define the ticker symbols
ticker_symbols = ['HINDUNILVR.NS', 'INFY.NS', 'ASIANPAINT.NS', 'HEROMOTOCO.NS']

# Set the start and end dates for the historical data
start_date = '2021-01-01'
end_date = '2024-05-01'

# Experimenting with different epochs and learning rates
epochs_list = [5, 10, 15]
learning_rates = [0.001, 0.005]

results = []

for ticker_symbol in ticker_symbols:
    # Fetch historical data from Yahoo Finance
    df = yf.download(ticker_symbol, start=start_date, end=end_date)
    
    # Use only the 'Close' column
    data = df[['Close']]
    
    # Convert the dataframe to a numpy array
    dataset = data.values
    
    # Scale the data
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(dataset)
    
    # Function to create dataset
    def create_dataset(data, time_step=1):
        x, y = [], []
        for i in range(len(data) - time_step - 1):
            x.append(data[i:(i + time_step), 0])
            y.append(data[i + time_step, 0])
        return np.array(x), np.array(y)
    
    best_mae = float('inf')
    best_model = None
    best_params = None

    for epoch in epochs_list:
        for lr in learning_rates:
            # Create dataset
            x, y = create_dataset(scaled_data, 60)
            x = x.reshape(x.shape[0], x.shape[1], 1)
            
            # Split data into training and testing sets
            train_size = int(len(x) * 0.8)
            x_train, x_test = x[:train_size], x[train_size:]
            y_train, y_test = y[:train_size], y[train_size:]
            
            # Build the LSTM model
            model = Sequential([
                LSTM(50, return_sequences=True, input_shape=(60, 1)),
                Dropout(0.2),
                LSTM(50, return_sequences=True),
                Dropout(0.2),
                LSTM(50, return_sequences=False),
                Dropout(0.2),
                Dense(25),
                Dense(1)
            ])
            # Compile the model
            model.compile(optimizer=Adam(learning_rate=lr), loss='mean_squared_error')
            
            # Train the model
            model.fit(x_train, y_train, batch_size=32, epochs=epoch, verbose=1)
            
            # Make predictions
            predictions = model.predict(x_test)
            predictions = scaler.inverse_transform(predictions)
            
            # Calculate performance metrics
            mae = mean_absolute_error(y_test, predictions)
            mse = mean_squared_error(y_test, predictions)
            rmse = sqrt(mse)
            
            print(f"Ticker: {ticker_symbol}, Epochs: {epoch}, LR: {lr}")
            print(f"MAE: {mae}, MSE: {mse}, RMSE: {rmse}")
            
            # Store results for plotting later
            results.append({
                'ticker': ticker_symbol,
                'params': (epoch, lr),
                'mae': mae,
                'mse': mse,
                'rmse': rmse
            })
            
            # Check if this model is better
            if mae < best_mae:
                best_mae = mae
                best_model = model
                best_params = (epoch, lr)
                
    # Print the best parameters and their performance
    print(f"Best Parameters for {ticker_symbol} - Epochs: {best_params[0]}, LR: {best_params[1]}")
    print(f"Best MAE: {best_mae}")

    # Save the best model
    best_model.save(f'best_lstm_model_{ticker_symbol}.h5')

    # Plot the data for the best model
    train_predict = best_model.predict(x_train)
    test_predict = best_model.predict(x_test)

    # Inverse transform predictions
    train_predict = scaler.inverse_transform(train_predict)
    test_predict = scaler.inverse_transform(test_predict)

    # Create DataFrames for plotting
    train = data[:train_size]
    test = data[train_size:train_size+len(test_predict)]

    test['Predictions'] = test_predict

    plt.figure(figsize=(16,8))
    plt.title(f'Model for {ticker_symbol} - Epochs: {best_params[0]}, LR: {best_params[1]}')
    plt.xlabel('Date', fontsize=18)
    plt.ylabel('Close Price INR (₹)', fontsize=18)
    plt.plot(train['Close'])
    plt.plot(test[['Close', 'Predictions']])
    plt.legend(['Train', 'Test', 'Prediction'], loc='lower right')

    # Save the figure in high resolution
    plt.savefig(f'stock_price_prediction_{ticker_symbol}_{best_params[0]}_{best_params[1]}.png', dpi=400, bbox_inches='tight')
    plt.show()

# Create a results dataframe
results_df = pd.DataFrame(results)
results_df_pivot = results_df.pivot_table(index=['ticker', 'params'], values=['mae', 'mse', 'rmse'])
results_df_pivot.to_csv('experiment_results.csv')

results_df_pivot
```
