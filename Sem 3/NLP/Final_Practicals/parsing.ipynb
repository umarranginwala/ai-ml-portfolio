```python
# Defining a context-free grammar (CFG)
grammar = nltk.CFG.fromstring("""
    S -> NP VP
    VP -> V NP | V NP PP
    V -> 'saw' | 'ate' | 'walks' | 'saw' | 'walked'
    NP -> 'John' | 'Mary' | 'Bob' | 'I' | Det N | Det N PP
    Det -> 'a' | 'an' | 'the' | 'my'
    N -> 'dog' | 'cat' | 'cookie' | 'park' | 'man' | 'park'
    PP -> P NP
    P -> 'in' | 'on' | 'by' | 'with'
""")
```

```python
# Function for parsing sentences using a recursive descent parser

def parsing(words):
    wrong_syntax = 1
    rd_parser = nltk.RecursiveDescentParser(grammar)
    for i in rd_parser.parse(words):
        s = i
        wrong_syntax = 0
        print('\nCorrect Grammar')
        return str(s)
    if wrong_syntax ==1:
        print('\nWrong Grammar')
        return str(s)
```
